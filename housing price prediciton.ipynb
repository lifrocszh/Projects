```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing data and data preparation

```python
df = pd.read_csv('1990 california housing prices.csv')        
df.head()
```

```python
set(df['ocean_proximity'])
```

```python
# Create dummy variables
dummies = pd.get_dummies(df['ocean_proximity'])

# Concatenate with original DataFrame
df = pd.concat([df, dummies], axis=1)
df.head()
```

```python
df = df.drop(['ocean_proximity'], axis=1)
df.head()
```

```python
for col in ['<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN']:
    df[col] = (df[col] == True).astype(int)
df.head()
```
